```python
# Import libraries for visualization
import pandas as pd
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')

# IMPORTANT: We are now loading the CLEANED data, not the raw data
file_path = '../data/processed/cleaned_apl_logistics.csv'
df = pd.read_csv(file_path)

print(f"Processed dataset loaded successfully! Rows: {df.shape[0]}")
```

```python
# 1. Delivery Classification Distribution
status_counts = df['delivery_classification'].value_counts().reset_index()
status_counts.columns = ['Delivery Status', 'Count']

# Create an interactive pie chart with specific color branding
fig1 = px.pie(status_counts, 
              values='Count', 
              names='Delivery Status',
              title='Distribution of Delivery Classifications',
              hole=0.4, # Makes it a modern donut chart
              color='Delivery Status',
              color_discrete_map={'Delayed': '#EF553B', 'On-time': '#00CC96', 'Early': '#636EFA'})

# Display the interactive plot
fig1.show()
```

```python
# 2. Average Delay by Shipping Mode
mode_delay = df.groupby('shipping_mode')['delay_gap'].mean().reset_index()

# Sort the values so the worst offenders are on the left
mode_delay = mode_delay.sort_values(by='delay_gap', ascending=False)

# Create a bar chart
fig2 = px.bar(mode_delay, 
              x='shipping_mode', 
              y='delay_gap',
              title='Average Delay Gap by Shipping Mode',
              labels={'shipping_mode': 'Shipping Mode', 'delay_gap': 'Avg Delay Gap (Days)'},
              color='delay_gap', 
              color_continuous_scale='Reds') # Darker red = worse delay

# Display the interactive plot
fig2.show()
```

```python
# 3. Delay Risk by Market
# We group by Market and calculate the average delay gap
market_delay = df.groupby('market')['delay_gap'].mean().reset_index()
market_delay = market_delay.sort_values(by='delay_gap', ascending=False)

# Create a bar chart for Regional Diagnostics
fig3 = px.bar(market_delay, 
              x='market', 
              y='delay_gap',
              title='Average Delivery Delay by Global Market',
              labels={'market': 'Global Market', 'delay_gap': 'Avg Delay Gap (Days)'},
              color='delay_gap',
              color_continuous_scale='Viridis') # Using a different color scale for variety

fig3.show()
```

```python
# 4. Delays across Customer Segments
segment_delay = df.groupby('customer_segment')['delay_gap'].mean().reset_index()
segment_delay = segment_delay.sort_values(by='delay_gap', ascending=False)

# Create a bar chart for Customer Segments
fig4 = px.bar(segment_delay, 
              x='customer_segment', 
              y='delay_gap',
              title='Average Delay by Customer Segment',
              labels={'customer_segment': 'Customer Segment', 'delay_gap': 'Avg Delay Gap (Days)'},
              color='customer_segment',
              color_discrete_sequence=px.colors.qualitative.Pastel)

fig4.show()
```
